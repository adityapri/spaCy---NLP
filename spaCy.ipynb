### Part of Speech Tagging (POS)

```python
text = "Aditya enjoys playing chess."
doc2 = nlp(text)
doc2
```

```python
for token in doc2:
  print(token.text, token.pos_, token.dep_)
```

```python
from spacy import displacy
displacy.render(doc2, style='dep')
```
